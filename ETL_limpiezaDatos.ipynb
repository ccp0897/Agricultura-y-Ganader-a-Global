# ETL - Limpieza y transformaciones de los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Carga de datos

```python
df_condiciones = pd.read_csv('datos/condiciones_climaticas.csv')
df_politicas = pd.read_csv('datos/politicas_agricolas.csv')
df_agri = pd.read_csv('datos/produccion_agricola.csv')
df_ganaderia = pd.read_csv('datos/produccion_ganadera.csv')
df_imag = pd.read_csv('datos/imagenes_metadata.csv')
df_precios = pd.read_csv('datos/precios_mercado.csv')
df_noticias = pd.read_csv('datos/noticias_agricolas.csv')
df_reportes = pd.read_csv('datos/reporte_plagas.csv')
```

Valores nulos

```python
#Valores nulos que existen
df_condiciones.isnull().sum()
```

Limpieza de nulos

```python
#Limpieza de nulos del dataframe de condiciones climáticas
df_condiciones["eventos_extremos"] = df_condiciones["eventos_extremos"].fillna("sin evento")
```

```python
df_condiciones.isnull().sum()
```

```python
#Limpieza de nulos del dataframe de políticas agrícolas
df_politicas['regulacion_multa_hectarea'] = df_politicas['regulacion_multa_hectarea'].fillna(0)
df_politicas['regulacion_porcentaje_minimo'] = df_politicas['regulacion_porcentaje_minimo'].fillna(0)
df_politicas['acuerdo_internacional_anio_firma'] = df_politicas['acuerdo_internacional_anio_firma'].fillna(0)
```

```python
df_politicas.isnull().sum()
```

Transformación de datos (fechas a datetime)

```python
# Crear fecha real en plagas mapeando meses en español a números si es necesario
meses_map = {'enero':1, 'febrero':2, 'marzo':3, 'abril':4, 'mayo':5, 'junio':6, 
             'julio':7, 'agosto':8, 'septiembre':9, 'octubre':10, 'noviembre':11, 'diciembre':12}
if df_reportes['Mes'].dtype == object: # Si viene en texto
    df_reportes['Mes_Num'] = df_reportes['Mes'].str.lower().map(meses_map)
else:
    df_reportes['Mes_Num'] = df_reportes['Mes']

# Convertir a datetime
df_reportes['Fecha_Completa'] = pd.to_datetime(
    df_reportes['Año'].astype(str) + '-' + 
    df_reportes['Mes_Num'].astype(str) + '-' + 
    df_reportes['Día'].astype(str), errors='coerce'
)
```

```python
df_reportes.info()
```

```python
df_precios['fecha'] = pd.to_datetime(df_precios['fecha'])
# Creamos columnas separadas porque los modelos de Machine Learning 
# necesitan variables numéricas independientes, no una fecha completa.
df_precios['anio_precio'] = df_precios['fecha'].dt.year
df_precios['mes_precio'] = df_precios['fecha'].dt.month
df_precios['dia_precio'] = df_precios['fecha'].dt.day
```

```python
df_precios.info()
```

```python
#Pasar datos limpios a csv para usarlos
df_reportes.to_csv('datos/reportes_plagas_limpios.csv', index=False)
df_precios.to_csv('datos/precios_mercado_limpios.csv', index=False)
df_condiciones.to_csv('datos/condiciones_climaticas_limpios.csv', index=False)
df_politicas.to_csv('datos/politicas_agricolas_limpios.csv', index=False)
```
